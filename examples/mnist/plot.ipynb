```python
import martin_util
import matplotlib.pyplot as plt
```

```python
# Plot training loss
setup_id = '1412_172546'
config = martin_util.txt_to_config_dict(setup_id)

df = martin_util.csv_to_dataframe(setup_id)
plt.title("Training loss " + setup_id)

epochs = int(config['epochs'])

for i in range(epochs):
    df_i = df[df['epoch'] == str(i+1)]
    df_i = df_i[df_i['train/test'] == 'train epoch']
    plt.plot(range(len(df_i)), df_i['loss/avg_loss'], label='epoch ' + str(i+1))

plt.legend()
```

```python
# Plot average test loss
plt.title(setup_id)
losses = []
df = martin_util.csv_to_dataframe(setup_id)
df_i = df[df['train/test'] == 'test']
print(df_i["loss/avg_loss"])
plt.scatter(range(len(df_i)), df_i['loss/avg_loss'], label='test loss')

```
